Downloads the OpenNeuro EEG dataset **ds002778** into `datasets/ds002778/`.

- **Dataset**: [ds002778 on OpenNeuro](https://openneuro.org/datasets/ds002778)
- **Source**: https://github.com/OpenNeuroDatasets/ds002778
- Download target will be `datasets/` directory, that is registered in `.gitignore` and will not be committed.

**Skip this notebook if `datasets/ds002778/` already exists.**

## 1. Configurations

```python
import openneuro
import os

DATASET_ID = 'ds002778'
TARGET_DIR = os.path.join('datasets', DATASET_ID)

os.makedirs(TARGET_DIR, exist_ok=True)
print(f'Target: {os.path.abspath(TARGET_DIR)}')
```

## 2. Downloading

```python
openneuro.download(dataset=DATASET_ID, target_dir=TARGET_DIR)
print('Download complete.')
```

## 3. Verifying

```python
# Verify structure
for root, dirs, files in os.walk(TARGET_DIR):
    level = root.replace(TARGET_DIR, '').count(os.sep)
    indent = '  ' * level
    print(f'{indent}{os.path.basename(root)}/')
    if level < 2:
        for f in files:
            print(f'{indent}  {f}')
```
